```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
df_survey = pd.read_csv('https://cocl.us/datascience_survey_data', index_col='Unnamed: 0')
```

```python
df_survey.head()
```

```python
df_survey.sort_values(by='Very interested', ascending=False, inplace=True)
df_survey
```

```python
df_survey_percentage=(100.*df_survey / 2233).round(2)
df_survey_percentage
```

```python
colors_list = ['#5cb85c', '#5bc0de', '#d9534f']
```

```python
ax=df_survey_percentage.plot(kind='bar', figsize=(20,8), width=0.8, color=colors_list)
plt.title("Percentage of Respondents' interest in Data science areas", size=16)
plt.legend(labels=df_survey_percentage.columns,fontsize= 14)
plt.xticks(fontsize=14)
for spine in plt.gca().spines.values():
    spine.set_visible(False)
plt.yticks([])

for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy() 
    ax.annotate('{:.2f}%'.format(height), (x, y + height + 0.01))
```
